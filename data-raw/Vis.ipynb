```python
selected_columns = [col for col in df2024.columns if 'Please select one option between each set of adjectives that best represents how you would rate your major department based on what you have seen and/or your own personal experience' in col]

df2024_q24=df2024[selected_columns]
df2024_q24.columns=[col.split(':')[1] for col in df2024_q24.columns]
df2024_q24=df2024_q24.join(df2024["What is your major? - Selected Choice"])
df2024_q24=df2024_q24.dropna()
df2024_q24=df2024_q24.groupby("What is your major? - Selected Choice").mean()
```

```python
import pandas as pd
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0)
dfc=df2024_q24-df2024_q24.min()/df2024_q24.max()-df2024_q24.min()
dfc['Cluster'] = kmeans.fit_predict(dfc)

merged_df = pd.DataFrame()

for cluster in set(dfc['Cluster']):

    cluster_df = dfc[dfc['Cluster'] == cluster]

    mean_values = cluster_df.mean(axis=0)
 
    new_index = ', '.join(cluster_df.index)

    merged_df = pd.concat([merged_df, pd.DataFrame(mean_values[:-1]).T.set_index(pd.Index([new_index]))])

# merged_df.index, merged_df.head()
```

```python
import matplotlib.pyplot as plt
from math import pi
# Adjusting the function to create a radar chart for each major, not each category
def create_radar_charts_for_majors(df):
    # Number of variables we're plotting.
    categories = list(df.index)
    N = len(categories)
    # df=df/df.sum()*100
    # df=df.mean()
    print(df)
    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Plot data and fill with color for each major
    for major in df.columns:
        fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

        # Draw one axe per variable + add labels for categories
        plt.xticks(angles[:-1], categories, color='black', size=8)
    
        # Draw ylabels
        ax.set_rlabel_position(0)
        # plt.yticks([200, 400, 600, 800], ["200", "400", "600", "800"], color="grey", size=7)
        plt.ylim(0, 5)
        
        # Plot data and fill with color
        values = df[major].tolist()
        values += values[:1]

        ax.plot(angles, values, color="green",linewidth=1.5, linestyle='solid', label=major)
        ax.fill(angles, values, color="green",alpha=0.1)

        # Add legend and title
        plt.title(major, size=15, color='black', y=1.1)
        # plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

        plt.show()

# Creating radar charts for each major
create_radar_charts_for_majors(df2024_q24.T)
```

```python
df2024_40=df2024['What are some things you like about your department?'].dropna()


import spacy
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

nlp = spacy.load("en_core_web_sm")


all_noun_2grams = []


for comment in df2024_40:

    doc = nlp(comment)
    nouns = [token.text for token in doc if token.pos_ == 'NOUN']
    # noun_2grams = [' '.join(nouns[i:i+2]) for i in range(len(nouns)-1)]
    all_noun_2grams.extend(nouns)


noun_2grams_text = ' '.join(all_noun_2grams)


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(noun_2grams_text)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import spacy
from collections import Counter
import pandas as pd


nlp = spacy.load("en_core_web_sm")


all_nouns = []


for comment in df2024_40:
    doc = nlp(comment)
    nouns = [token.text.lower() for token in doc if token.pos_ == 'NOUN']  #
    all_nouns.extend(nouns)


noun_counts = Counter(all_nouns)


top_nouns = noun_counts.most_common(5)


print(top_nouns)
```

```python
df2024_41=df2024['Any suggestions for things your department should do in the future?'].dropna()


import spacy
from collections import Counter


nlp = spacy.load("en_core_web_sm")



suggestive_phrases = []


for text in df2024_41:
    doc = nlp(text)
    for token in doc:

        if token.pos_ == "VERB":

            for child in token.children:
                if child.dep_ in ["dobj", "attr", "nsubjpass"]:

                    phrase = token.lemma_ + " " + child.lemma_
                    suggestive_phrases.append(phrase.lower())


phrase_counts = Counter(suggestive_phrases)
top_suggestive_phrases = phrase_counts.most_common(5)

print(top_suggestive_phrases)
```

```python
import spacy
from spacy.tokens import Token


nlp = spacy.load("en_core_web_sm")



def extract_subtree_text(token):
    return " ".join([child.text for child in token.subtree])


full_phrases = []

for text in df2024_41:
    doc = nlp(text)
    for token in doc:

        if token.lemma_ == "provide" and token.pos_ == "VERB":

            full_phrase = extract_subtree_text(token)
            full_phrases.append(full_phrase.lower())

print(full_phrases)
```
